## **Simple Plot lines**

```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-white')
fig = plt.figure()
ax = plt.axes()
```

figure (an instance of class plt.Figure) can be thought of as a single container that contains all the objects representing axes,graphics, text and labels.

axex(an instance of the class plt.Axes) is a bounding box with ticks,grids,and labels which will eventually contain the plot elements that make up our visualizations

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

better use plt.plot directly instead

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.grid()
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code(rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # hex code
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple,values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names supported
```

If we dont' specify the color matplotlib will cycle through a set of default colors for multiple lines

```python
plt.grid()
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

";" at the end is intentional: it suppresses the textual representation of the plot from the output

```python
plt.grid()
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

It is also possible to combine the color and linestyle

```python
plt.grid()
plt.plot(x, x + 0, '-g')   # solid green
plt.plot(x, x + 1, '--c')  # dashed cyan
plt.plot(x, x + 2, '-.k')  # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

One can display the axis in reverse by simply reversing the order of the arguments

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.axis('tight');
```

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.axis('equal');
```

```python
plt.grid()
plt.plot(x,np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.grid()
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')
plt.legend();
```

## **Simple Scatter Plots**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.grid()
plt.plot(x,y,'o',color='black');
```

```python
plt.grid()
plt.plot(x,y,"-ok");
```

```python
plt.scatter(x,y,marker='o')
plt.grid();
```

plt.scatter has control over properties of each individual point(size,face color,edge color, etc.) which can be controlled individually or mapped to data.

*alpha* is used to control transparency

```python
rng = np.random.default_rng(0)
x = rng.normal(size=100)
y = rng.normal(size=100)
colors = rng.random(100)
sizes = 1000 * rng.random(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar();  # show color scale
```

we can use the color, position and size argument to convey more information, taking the example of the iris dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0],features[1],alpha=0.6,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.grid();
```

```python
print(iris.feature_names)
```

the position of the instance tells the sepal length and width,(x,y), the color tells the species of the iris flower and the size of the instance corresponds to the petal length, but remember trying to convey more information can make it more confusing to interpret so try to strike a balance before the visuals become overpowering

plt.plot works faster than plt.scatter on larger dataset as scatter renders each instance seperately while plot assumes that the markers for each point to be identical 

### **Errorbars**

For any scientific measurement there is always a uncertainity maybe from the human or from the equipment so it is necessary to show that as well.

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='.g')
plt.grid()
plt.show()
```

fmt is the format of plotting the errorbar and yerr is the uncertainity

```python
plt.errorbar(x,y,yerr=dy,fmt='o',color='black',ecolor='lightgray',elinewidth=3,capsize=0)
plt.grid();
```

sometimes we want to the show errors as continuous function instead of discrete values for that we combine plt.plot and plt.fill_between

```python
from sklearn.gaussian_process import GaussianProcessRegressor
model = lambda x: x*np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

gp = GaussianProcessRegressor()
gp.fit(xdata[:,np.newaxis],ydata)
xfit = np.linspace(0,10,100)
yfit,dyfit = gp.predict(xfit[:,np.newaxis],return_std=True)
```

```python
plt.plot(xdata,ydata,'or')
plt.plot(xfit,yfit,'-',color='gray')
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,color='gray',alpha=0.2)
plt.xlim(0,10)
plt.grid();
```

## **Density and Contour Plots**
Used to display 3 dimensional data in two dimensions using contours or color-coded regions

plt.contour for contour plots

plt.contourf for filled contur plots 

plt.imshow for showing images


```python
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
```

plt.contour takes x,y,z where x and y are the positions and y is the color 

to conver one=dimensionl arrays to two-dimensional grids we use np.meshgrid function

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contour(X,Y,Z,colors='black');
```

negative values are represented by dashed lines and solid ones by solid lines, we can also specify a color map

we use 20 to tell that we want more lines to be drawn at 20 equally spaced intervals

```python
plt.contour(X,Y,Z,20,cmap='RdGy');
```

```python
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar();
```

from the colorbar we can judge that the black regions are the peaks and red values are the valleys

but the plot looks splotcy; the colors are discrete than continuous we can overcome them by using imshow

```python
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy',interpolation='gaussian',aspect='equal')
plt.colorbar();
```
